```python
tdata2 = np.array([0.5,0.25,0.75,1,1.25,1.5,1.75,2,2.25,2.5,4,7.5])
discount = []

for j in range(12):
    ydata2 = (coef[j][0] + coef[j][1]*tdata2)*np.exp( - coef[j][2]*tdata2) + coef[j][3]
    discountfactor = []
    for i in range(len(tdata2)):
        discountfactor.append(1/(ydata2/100+1)[i] ** tdata2[i])
    discount.append(discountfactor)
discountfactor
```

```python
discountdata = pd.DataFrame(discount, columns = tdata2)
swaprate = []

for i in range(12):
    w = (1 - discountdata.loc[i][7.5])/sum(discountdata.loc[i])*100
    swaprate.append(w)
    
swaprate
```

Compute MTM values of the swaps (time series of swaps)

```python
MTMvalues = pd.DataFrame(discount)[0] * swaprate
MTMvalues
```
